Bài 4: Viết chương trình in ra hình tam giác cân có độ cao h được nhập từ bàn phím

```python
h=int(input())
for i in range(1,h+1):
    if i==1:
        print(' '*(h-i)+'*')
    elif i==h:
        print('*'*((i-1)*2+1))
    else:
        print(' '*(h-i)+'*'+' '*((i-1)*2-1)+'*')
```

Bài 6: Nhập vào số nguyên dương a, in ra tất cả các ước dương của a 

```python
a=int(input())
for i in range(1,int(a**0.5)+1):
    if a%i==0:
        print(i, end=' ')

for i in range(int(a**0.5),0,-1):
    if a%i==0:
        uoc=a//i
        if uoc!=i:
            print(uoc,end=' ')
```

Bài 7: Nhập vào số nguyên dương a, đếm số ước dương của a

```python
a=int(input())
count=0
for i in range(1,int(a**0.5)+1):
    if a%i==0:
        count+=1
        if i!=(a/i):
            count+=1
print(count)
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.

```python
a=int(input())
sum=0
for i in range(1,int(a**0.5)+1):
    if a%i==0:
        sum+=i
        if i!=(a/i) and (a/i)!=a:
            sum+=(a//i)

if sum==a:
    print('Yes')
else:
    print('No')
```

Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k) lớn nhất nhưng nhỏ hơn n

```python
n=int(input())
sum=0
k=0
while sum+k+1<n:
    k+=1
    sum+=k
print(k)
```

Bài 13: Nhập vào số nguyên dương A, tìm n nhỏ nhất sao cho

1 + 1/2 + 1/3 + 1/4 + ... + 1/n > A 


```python
A=int(input())
sum=0
n=0
while sum<=A:
    n+=1
    sum+=(1/n)
print(n)
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập

```python

n=int(input())
max=0
min=0
while n!=-1:
    n=int(input())
    if n>max:
        max=n
    if n<min:
        min=n

print(max,min)


```

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá A

```python
A=int(input())
f1=0
f2=1
sum=0
while sum<A:
    sum=f1+f2
    f1=f2
    f2=sum
print(f1)
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố

```python
n=int(input())
count=0

for i in range(2,n+1):
    for j in range(2,int(i**0.5)+1):
        if i%j==0:
            break
    else:
        count+=1
print(count)


```
