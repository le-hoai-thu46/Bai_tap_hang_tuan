1. Kiểm tra số chẵn lẻ: Viết chương trình nhập một số nguyên, kiểm tra và in ra số đó
là số chẵn hay số lẻ.

```python
n=int(input('nhập một số nguyên:'))
if n%2==0:
    print(f"{n} là số chẵn")
else:
    print(f'{n} là số lẻ')
```

4. Kiểm tra tuổi đủ điều kiện bầu cử: Viết chương trình nhập năm sinh của một
người, kiểm tra xem người đó có đủ 18 tuổi để đi bầu cử hay không. 


```python
n=int(input('nhập năm sinh của bạn:'))
if 2026-n>=18:
    print('bạn đủ 18 tuổi để đi bầu cử')
else:
    print('bạn chưa đủ 18 tuổi để đi bầu cử')
```

5. So sánh 2 số: Viết chương trình nhập vào hai số, in ra số lớn hơn hoặc thông báo
nếu hai số bằng nhau

```python
a,b=map(float,input().split(','))
if a!=b:
    print('số lớn hơn là:',max(a,b))
else:
    print('đây là hai số bằng nhau')
```

6. Kiểm tra ký tự nhập vào là chữ hay số: Viết chương trình nhập một ký tự, kiểm
tra xem ký tự đó là chữ cái hay số. 

```python
n=input('nhập một thứ bất kì:')
#for c in n:
if n.isdigit():
    print(n,'là số')
elif n.isalpha():
    print(n,'là chữ')
else:
    print(n,'kí tự khác')
```

7. Bài tập 5: Tính điểm học lực
Nhập điểm (0-10), phân loại học lực theo quy tắc: >=8: Giỏi; >=6.5 và <8: Khá; >=5 và <6.5: Trung bình; <5: Yếu

```python
n=float(input('nhập điểm:'))
if n>=8:
    print('giỏi')
elif n>=6.5:
    print('khá')
elif n>=5:
    print('trung bình')
else:
    print('yếu')
```

8. Kiểm tra năm nhuận: Viết chương trình nhập năm, kiểm tra xem năm đó có phải
năm nhuận hay không. (Năm nhuận nếu chia hết cho 400 hoặc chia hết cho 4
nhưng không chia hết cho 100) 

```python
n=int(input('nhập năm:'))
if n%4==0 or n%400==0:
    if n%100!=0:
        print(n,'là năm nhuận')
    else:
        print(n,'không phải là năm nhuận')
else:
    print(n,'không phải là năm nhuận')
```

9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case 

```python
n=int(input('nhập một chữ số:'))
match n:
    case 0:print('Không')
    case 1:print('Một')
    case 2:print('Hai')
    case 3:print('Ba')
    case 4:print('Bốn')
    case 5:print('Năm')
    case 6:print('Sáu')
    case 7:print('Bảy')
    case 8:print('Tám')
    case 9:print('Chín')
```

10. Kiểm tra điểm thi xem có qua môn hay phải học lại: Nhập điểm thi, nếu điểm >=
4 thì in "Qua môn", ngược lại in "Học lại". 


```python
n=float(input('nhập điểm của bạn:'))
if n>=4:
    print('Qua môn')
else:
    print('Học lại')
```

11. Kiểm tra năm sinh và tính tuổi: Nhập năm sinh, tính tuổi hiện tại (giả sử năm
hiện tại là 2025), và kiểm tra xem người đó có đủ 18 tuổi hay chưa. 

```python
n=int(input('nhập năm sinh của bạn:'))
age=2025-n
if age>=18:
    print(f'Tuổi của bạn là: {age}, bạn đủ 18 tuổi')
else:
    print(f'Tuổi của bạn là: {age}, bạn chưa đủ 18 tuổi')
```
