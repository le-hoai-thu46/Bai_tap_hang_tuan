Các quy tắc cần cài đặt:

    1. Giá mở cửa (0–0.8 km):
    - Xe 4 chỗ: 11,000 VND
    - Xe 7 chỗ: 12,000 VND

    2. Giá theo km sau 0.8 km:
    - Xe 4 chỗ: từ 0.8→20 km ơnh 13,000 VND/km; trên 20 km tính 10,500 VND/km
    - Xe 7 chỗ: từ 0.8→20 km ơnh 14,000 VND/km; trên 20 km tính 12,000 VND/km

    chờ).

    5. Điều kiện đặc biệt:
    - Nếu km > 30 và không phải ban đêm, áp dụng giảm 10% trên tổng tiền.
    - Nếu tổng tiền cuối cùng < 50,000 nhưng km > 5, đặt tổng tiền bằng 50,000 VND (mức tối thiểu).

16. Viết hàm trả về góc ∠ABC theo đơn vị độ khi biết tọa độ các điểm A(x1,y1), B(x2,y2), C(x3,y3). 

```python
import math
def goc(x1,y1,x2,y2,x3,y3):
    vba=[(x1-x2),(y1-y2)]
    vbc=[(x3-x2),(y3-y2)]
    ba=math.hypot(vba[0],vba[1])
    bc=math.hypot(vbc[0],vbc[1])
    ts=vba[0]*vbc[0]+vba[1]*vbc[1]
    cosb=ts/(ba*bc)
    goc=math.acos(cosb)
    return math.degrees(goc)
goc(1,5,2,2,5,3)
```

và False nếu ngược lại.
- Đầu vào:
- quad: danh sách hoặc bộ gồm đúng 4 đỉnh [(x1,y1), (x2,y2), (x3,y3), (x4,y4)] theo thứ tự (tứ giác có thể lồi hoặc lõm,
nhưng không tự cắt).
- p: một điểm (x, y).
- Đầu ra: giá trị boolean.
- Gợi ý: Cài đặt phép kiểm tra điểm trong đa giác tổng quát (ví dụ: ray casting), đồng thời tính cả trường hợp điểm nằm
trên biên là ở trong.

```python
def point_in_quadrilateral(quad, p):
    cnt=0
    for i in range(4):
        x1,y1 = quad[i]
        x2,y2 = quad[(i+1) % 4]

        if ((p[1] - y1) * (x2 - x1)) == ((p[0] - x1)*(y2 - y1)):
            if min(x1, x2) <= p[0] <= max(x1, x2) and min(y1, y2) <= p[1] <= max(y1, y2):
                return 'Tren bien'
        if (y1 > p[1]) != (y2 > p[1]):
            x_giao=x1+((p[1]-y1)*(x2-x1))/(y2-y1)
            if x_giao > p[0]:
                cnt += 1

    if cnt % 2 != 0:
        return 'Trong'
    else:
        return 'Ngoai'
point_in_quadrilateral([(0, 0), (0, 4), (4, 4), (4, 0)], [0,2])
```

- Đầu vào: chuỗi s.
- Đầu ra: một trong các giá trị {'int', 'float', 'neither'}.
- Quy tắc:
- Cho phép có dấu + hoặc - ở đầu.
- Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100').
- Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14').
- Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này. 

```python
def detect_number_type(s):
    if s.startswith('-'):
        s=s[1:]
    if '.' not in s:
        if s.isdigit():
            return 'int'
        else:
            return 'neither'
    else:
        if s.startswith('.'):
            return 'neither'
        try:
            float(s)
            return 'float'
        except ValueError:
            return 'neither'
detect_number_type('.415')
            
```

32. Yêu cầu: Viết hàm is_right_triangle(a, b, c) trả về True nếu a, b, c có thể là ba cạnh của một tam giác vuông, ngược lại trả
về False.
- Đầu vào: các số dương a, b, c.
- Đầu ra: giá trị boolean.
- Lưu ý: Kiểm tra bất đẳng thức tam giác và định lý Pythagore. Dùng sai số nhỏ khi so sánh số thực. 


```python
def is_right_triangle(a, b, c):
    a, b, c = sorted([a, b, c])
    if a <=0:
        return False
    else:
        if c < a+b and abs((a**2 + b**2) - (c**2)) < 1e-3:
            return True
        return False
is_right_triangle(1, 1, 1.4142)
```

- Đầu ra: kết quả số học.
- Yêu cầu thêm:
- Nếu op là '/' và b == 0, phát sinh ValueError.
- Nếu op không hợp lệ, phát sinh ValueError. 


```python
def compute(a, b, op):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        if b == 0:
            return 'ValueError'
        else:
            return a / b
compute(2, 0, '/')
    
    
```

cho dividend = divisorq + r với 0 <= |r| < |divisor| (dùng ngữ nghĩa của Python với // và %).

- Đầu ra: bộ hai số nguyên (q, r).
- Ví dụ: divide_with_remainder(5, 3) -> (1, 2).

37. Viết hàm trả về tam giác vuông cân cao n bằng ký tự ''.
Dòng thứ i (đánh số từ 1) có i dấu '', các dòng nối với nhau bằng ký tự xuống dòng.
Ví dụ với n = 4 

```python
def ve_tam_giac(n):
    line_mid = ''
    for i in range(1, n + 1):
        line = '*' * i
        if i < n:
            line_mid += line + '\n'
        else:
            line_mid += line
    return line_mid
print(ve_tam_giac(4))
```

43. Viết hàm tính diện tích tam giác có ba cạnh a, b, c (dùng công thức Heron). Phát sinh ValueError nếu ba cạnh không thể
tạo thành tam giác. 

```python
def dien_tich_tam_gaic(a, b, c):
    a, b, c = sorted([a, b, c])
    if a > 0 and c <= a + b:
        p = (a + b + c) / 2
        dt = (p * (p - a) * (p - b) * (p -c)) ** 0.5
        return dt
    else:
        return 'ValueError'
dien_tich_tam_gaic(1, 1, 2)
```

44. Viết hàm trả về tích các số nguyên từ 1 đến n (giai thừa). Với n=0, trả về 1

```python
def tich(n):
    if n == 0:
        return 1
    gt = 1
    while n > 0:
        gt *= n
        n -= 1
    return gt
tich(5)
```

45. Viết hàm trả về số lượng số nguyên dương k sao cho 1 <= k < n và k > (n - 1) / 2. (Ví dụ: n=9 -> 4). 


```python
def so_luong_k(n):
    cnt = 0
    for i in range(1, n + 1):
        if 1 <= i and i < n and i > ( n - 1) / 2:
            cnt += 1
    return cnt
so_luong_k(9)
```
