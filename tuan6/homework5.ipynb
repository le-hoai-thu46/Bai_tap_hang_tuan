W5A20. Viết hàm detect_number_type(s) trả về 'int' nếu s là số nguyên, 'float' nếu là số thực, ngược lại 'neither'.

```python
def  detect_number_type(s):
    try:
        if '.' not in s:
            int(s)
            return 'int'
        else:
            if s[0]=='.' or s[-1]=='.':
                return 'neither'
            else:
                float(s)
                return 'float'
            
    except ValueError:
        return 'neither'

detect_number_type('123.34')
```

W5A24. Viết hàm are_collinear(x1,y1,x2,y2,x3,y3) kiểm tra ba điểm có thẳng hàng hay không

```python
def  are_collinear(x1,y1,x2,y2,x3,y3):
    ab=[(x2-x1), (y2-y1)]
    ac=[(x3-x1), (y3-y1)]
    return (ab[0]*ac[1])==(ab[1]*ac[0])
are_collinear(1,2,2,4,3,6)
```

Viết hàm point_in_quadrilateral(quad, p) kiểm tra điểm p có nằm trong hoặc trên biên của tứ giác quad hay
không. Trong đó: quad là một tứ giác (4 đỉnh – 4 điểm), p là một điểm cần kiểm tra. VD: quad = [(0,0), (4,0), (4,3),
(0,3)], p = (2,1)


```python
def point_in_quadrilateral(quad, p):
    def tich_co_huong(a,b,c):
        return ((b[0]-a[0])*(c[1]-a[1])-(b[1]-a[1])*(c[0]-a[0]))
    ktra_phia=[]
    for i in range(4):
        canh_dau=quad[i]
        canh_cuoi=quad[(i+1)%4]
        dau=tich_co_huong(canh_dau, canh_cuoi, p)

        if dau!=0:
            ktra_phia.append(dau>0)
    for i in ktra_phia:
        if not i:
            return "Điểm nằm ngoài tứ giác"
        return "Điểm nằm trong hoặc trên biên tứ giác"
        
quad = [(0,0), (4,0), (4,3), (0,3)]
p = (2,1)

print(point_in_quadrilateral(quad, p))

```

W5A28. Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
import math
def goc(x1,y1,x2,y2,x3,y3):
    vt_ba=[(x1-x2),(y1-y2)]
    vt_bc=[(x3-x2),(y3-y2)]
    ba=math.hypot(vt_ba[0],vt_ba[1])
    bc=math.hypot(vt_bc[0],vt_bc[1])
    tu_so=vt_ba[0]*vt_bc[0]+vt_ba[1]*vt_bc[1]
    goc=math.acos((tu_so)/(ba*bc))
    return math.degrees(goc)
goc(1,5,2,2,5,3)

```

W5A30. Viết hàm tính hệ số nhị thức C(n, k).


```python
def gt(n):
    gt=1
    while n:
        gt*=n
        n-=1
    return gt

def C(n,k):
    return (gt(n)/(gt(k)*gt(n-k)))
C(5,3)
```

W5A34.
Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau 


```python
def  sum_1_to_n(n):
    sum=0
    
    for i in range(1,n+1):
        so=i
        lap=[False]*10
        only=True

        while so>0:
            chuso=so%10
            if lap[chuso]==True:
                only=False
                break

            lap[chuso]=True
            so//=10
        if only:
            sum+=i
    return sum
sum_1_to_n(12)
```
