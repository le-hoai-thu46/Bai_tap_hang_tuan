 Bài 2: Một khu đất hcn có chiều rộng a, chiều dài b. Một khu vui chơi hình tròn tại vị trí trung tâm khu đất. Nhập vào a, b. Hãy tính phần diện tích còn để trồng cây xung quanh, pi = 3.14, lấy kết quả đến 2 chữ số phần thập phân

```python
a,b=map(float,input().split(','))
hcn=a*b
pi=3.14
r=min(a,b)/2
htron=r**2*pi
s=hcn-htron
print(f'diện tích trồng cây:{s:.2f}')
```

 Bài 3: Viết chương trình nhập vào một “kí tự” c, nếu nó là chữ hoa thì in ra chữ thường tương ứng và ngược lại nếu nó là chữ thường thì in ra chữ hoa tương ứng của nó. (xem trước cấu trúc lựa chọn: if….else, ma)

```python
n=input('nhập một kí tự bất kì:')
if len(n)==1:
    if n.isupper():
        print('kí tự được chuyển sang in thường:',n.lower())
    elif n.islower():
        print('kí tự được chuyển sang in hoa:',n.upper())
    else:
        print("Kí tự nhập vào không phải là chữ cái.")
else:
    print("Vui lòng chỉ nhập một ký tự duy nhất.")
```

 Bài 4: Viết chương trình nhập vào một “kí tự”, Kiểm tra xem c có phải là kí tự chữ cái không (hoa hoặc thường). Nếu đúng thì in ra “{c} là kí tu alphabet”, nếu sai in ra “{c} không phải là kí tự alphabet.

```python
c=input('nhập một kí tự:')
if len(c)==1:
    if c.isalpha():
        print(c,' là kí tự alphabet')
    else:
        print(c,' không phải là kí tự alphabet')
else:
    print('không đúng định dạng, nhập lại')
```

Bài 5: Nhập vào chữ cái hoa A, in ra chữ cái thường liền trước chữ cái a thường tương
ứng của A. Xét TH khi c = ‘A’ là đặc biệt

```python
A=input('nhập một chữ cái in hoa:')
A=A.lower()
if A=="z":
    print('chữ cái thiếp theo là: a')
else:
    print('chữ cái tiếp theo: ',chr(ord(A)+1))
#ord biến kí tự thành số, chr chuyển ngược lại
```

Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 


```python
a,b,c=map(float,input().split())
if a>0 and b>0 and c>0:
    if a+b>c and a+c>b and b+c>a:
        p=(a+b+c)/2
        s=(p*(p-a)*(p-b)*(p-c))**0.5
        print(f'diện tích tam giác: {s:.1f}')
    else:
        print('không phải 3 cạnh tam giác')
else:
    print('không phải 3 cạnh của tam giác')
```
