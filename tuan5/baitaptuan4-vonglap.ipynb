W4A2. VIết chương trình nhập vào một số tới khi được một số nguyên dương thì thôi.
Kiểm tra số này có phải là số nguyên tố hay không? 

```python
while True:
    so=int(input())
    if so>0:
        break

if so<2:
    print('No')
else:
    isprime=True
    for i in range(2,int(so**0.5)+1):
        if so%i==0:
            isprime=False
            break

    if isprime:
        print('Yes')
    else:
        print('No')

```

W4A5. Hãy viết một chương trình nhận vào một số nguyên dương n và một dãy gồm n số
nguyên. Nếu n số nguyên đó có số 42 thì in ra màn hình dòng chữ "I′ve found the
meaning of life!", ngược lại in ra dòng chữ "It′s a joke!"

```python
n=int(input())
while True:
    day_n=list(map(int,input().split()))
    if len(day_n)==n:
        break

for i in day_n:
    if i==42:
        print("I′ve found the meaning of life!")
        break
else:
    print("It′s a joke!")

```

W4A6. Nhập hai số nguyên dương a, b (a ≤ b). Hãy tính tổng các số nguyên tố trong đoạn
[a, b]. Mỗi dòng là một bộ test (a b)

```python

t=2
def snt(n):
    for i in range(2, int(n**0.5)+1):
        if n % i== 0:
            return False
    return n > 1

while t > 0:
    a, b = map(int, input().split())
    if min(a,b)>0 and a<=b:
        sum = 0
        for i in range(a, b + 1):
            if snt(i):
                sum += i
        print(sum)
        
        t -= 1
```

W4A7. Nhập vào số nguyên dương n (n ≥ 2). Hãy in ra ước số nguyên tố lớn nhất của n."

W4A8. Nhập số nguyên dương n. Lặp lại phép toán n = n + reverse(n) cho đến khi n là số
đối xứng (palindrome). In ra số bước và giá trị palindrome thu được. 

```python
def rev(n:int):
    n_rev=0
    while n:
        n_rev=n_rev*10+n%10
        n//=10
    return n_rev

def palindrome(n):
    return n!=rev(n)

n=int(input())
cnt=0

while palindrome(n):
    n+=rev(n)
    cnt+=1
print(n,cnt)
```

W4A10. Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số nhỏ nhất.


```python
def collazt(n:int):
    if n==1:
        return 0
    cnt=0
    while(n!=1):
        if n%2==0:
            n//=2
        else:
            n=n*3+1
        cnt+=1
    return cnt

n=int(input())
val_max=1
val_collazt=0
for x in range(1,n+1):
    if val_collazt<=collazt(x):
        val_max=x
        val_collazt=collazt(x)
print(val_max, val_collazt+1)
```

W4A13. Nhập hai số từ bàn phím và kiểm tra xem chúng có phải là cặp số thân thiết không. Nếu có, in ra màn hình "true", ngược lại, in ra "false" 

```python
def uoc(n):
    sum=0
    for i in range(1,int(n**0.5)+1):
        if n%i==0:
            sum+=i
            if i!=n/i and (n/i)!=n:
                sum+=n/i
    return sum

a,b=map(int,input().split())
tong_uoc_a=uoc(a)
tong_uoc_b=uoc(b)

if tong_uoc_a==b and tong_uoc_b==a:
    print('True')
else:
    print('False')

```
