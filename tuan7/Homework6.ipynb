W6A16. Nhập vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải


```python
n=int(input())
m=int(input())
mt=[]
for i in range(n):
    row=[]
    for j in range(m):
        val=int(input())
        row.append(val)
    mt.append(row)
for row in mt:
    for val in row:
        print(f'{val:4d}', end='')
    print()
        
    
```

W6A17. Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
đường chéo phụ của mảng

```python
n=int(input())
l=list(map(int,input().split()))
mt=[l[i*n:(i+1)*n] for i in range(n)]

for i in range(n):
    for j in range(n):
        if j == i or j == n - 1 - i:
            print(mt[i][j], end=" ")
        else:
            print(" ", end=" ")
    print()
```

W6A18. Nhâ ̣p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên ̣
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột
k (hàng k) 


```python
n=int(input())
m=int(input())
k=int(input())

if k>m or k>n:
    print('invalid')
else:
    l=list(map(int, input().split()))
    mt=[l[i*m:(i+1)*m] for i in range(n)]
    x=sum(mt[k-1])
    y=sum(row[k-1] for row in mt)
    print(y)
    print(x)
```

W6A21. Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
l=input().split()
d={}

for i,v in enumerate(l):
    if v not in d:
        d[v]=[]
    d[v].append(i)
for key in d:
    d[key]=tuple(d[key])
print(d)
```

W6A22. Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
trong list

```python
l=list(map(int, input().split()))
maxi=0
mini=0

for i in range(len(l)):
    if l[i]>=l[maxi]:
        maxi=i
    if l[i]<=l[mini]:
        mini=i
print(maxi)
# print(mini)
```

W6A24. Viêết hàm đẩy tất cả các số 0 về đầu (về cuối), các số khác vẫn đúng vị trí
trước sau: 0 2 3 2 4 0 0 0 9 => 0 0 0 2 3 2 4 9


```python
l=list(map(int, input().split()))
cnt=0
s=[]
for i in l:
    if i==0:
        cnt+=1
    else:
        s.append(i)
print([0]*cnt+s)
```

W6A26. Tính tổng các phần tử cực đại địa phương trong một danh sách, không tính 2
phần tử đầu và cuối


```python
l=list(map(int, input().split(',')))
sum=0
for i in range(1,len(l)-1):
    if l[i-1]<l[i]>l[i+1]:
        sum+=l[i]
print(sum)
```

W6A28. Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
l=list(map(int, input().split(',')))
kt_tang=True
kt_giam=True
for i in range(len(l)-1):
    if l[i]<l[i+1]:
        kt_giam=False
    elif l[i]>l[i+1]:
        kt_tang=False
    else:
        kt_tang=False
        kt_giam=False
    
if kt_tang:
    print('Tang')
elif kt_giam:
    print('Giam')
else:
    print('Vo danh')
```
