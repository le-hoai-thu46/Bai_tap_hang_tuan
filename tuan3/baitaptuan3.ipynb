W3A1. Viết một chương trình nhập vào một số nguyên n và in ra các chữ số đảo ngược nhau

```python
n=(input('Nhập một số nguyên:'))
if n[0]=='-':
    print('-',n[:0:-1])
else:
    print(n[::-1])
```

```python
n=int(input())

if n>0:
    n_rev=0
    for _ in range(n+1):
        if n==0:
            break
            #điều kiện để dừng
        n_rev=n_rev*10+n%10
        n//=10
    print(n_rev)
else:
    print('Ko sử lí được')
    
```

```python
n=int(input())
n_rev=0
if n>0:
    while n:
        n_rev=n_rev*10+n%10
        n//=10
        #nếu n//=10 lên trc ví dụ 12345//10=1234, mất số cuối-phần dư
    print(n_rev)
else:
    print('ko xử lí được')
```

```python
a,b,c=map(int,input().split())
if a>0 and b>0 and c>0:
    if a+b>=c and b+c>=a and a+c>=b:
        if a==b==c:
            print('Tam giác đều')
        elif a==b or b==c or a==c:
            print('Tam giác cân')
    else:
        print('Tam giác thường')
else:
    print('Không phải tam giác')
```

W3A13. Nhập vào số kWh điện tiêu thụ. Tính tiền điện theo các bậc sau:
- 0-50 kWh: 1500đ/kWh
- 51-100 kWh: 2000đ/kWh
- 100 kWh: 3000đ/kWh 

```python
n=float(input('Nhập số điện:'))
sum=(
    min(n,50)*1500+
    min(max((n-50),0),50)*2000+
    min(max((n-100),0),100)*3000)
print(sum)
```
