Bài 3 : Tính tam giác: Nhâp vào ba số a,b,c bất kì. Kiểmtra xem nó có thể làđộ dài ba cạnh hay không, nếu không thì in ra màn hình 'Khong la ba canh cua mot tam giac'. Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a,b,c=map(float,input().split())
a,b,c=sorted([a,b,c],reverse=True)
a_binh=a**2
b_binh=b**2
c_binh=c**2
if a>0 and b>0 and c>0 and b+c>a:
    if a==b==c:
        print('Tam giác đều')
    elif a_binh-(b_binh+c_binh)<0.0001:
        print('Tam giác vuông')
    elif a==b or a==c or b==c:
        print('Tam giác cân')
    else:
        print('Tam giác thường')
else:
    print('Không phải là ba cạch của tam giác')
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a,b,c=map(float,input().split())
if a!=0:
    delta=b**2-4*a*c
    if delta<0:
        print('Phương trình vô nghiệm')
    elif delta==0:
        print('Phương trình có nghiệm kép:',-b/(2*a))
    else:
        x1=(-b-delta**0.5)/(2*a)
        x2=(-b+delta**0.5)/(2*a)
        print('phương trình có hai nghiệm phân biệt:',x1,x2)
else:
    print('Không phải phương trình bậc 2')
```

Bài 8: Giải hệ phương trình tuyến tính:

ax + by = m

cx + dy = n

```python
a,b,c,d,m,n=map(float,input().split())
if a==c and b==d and m==n:
    print('Hệ phương trình có vô số nghiệm')
elif a==c and b==d and m!=n:
    print('Hệ phương trình có vô nghiệm')
else:
    x=(b*n)/(b*c+d*m-d*a)
    y=(m-a*x)/b
    print('nghiệm hệ phương trình:',x,y)
```

Bài 10: Nhập vào tâm và bán kính của một đường tròn. Sau đó nhập vàomộtđiểm A(x, y) bất kì và kiểm tra xem nó có thuộc đường tròn hay không?

```python
r=float(input())
a,b=map(float,input().split())
x,y=map(float,input().split())
if (x-a)**2+(y-b)**2==r**2:
    print('Điểm A thuộc đường tròn')
elif (x-a)**2+(y-b)**2>r**2:
    print('Điểm A nằm ngoài đường tròn')
else:
    print('Điểm A nằm trong đường tròn')
```
